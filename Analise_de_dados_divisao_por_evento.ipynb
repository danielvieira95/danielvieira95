Variaveis para calcular a diferença entre as linhas e filtrar por evento

```python
df['dif_p2_ciot_in'] = 0
df['dif_p2_bpc_in'] = 0
df['dif_p2_ciot_out'] = 0
df['dif_p2_bpc_out'] = 0
df['dif_p3_ciot_in'] = 0
df['dif_p3_bpc_in'] = 0
df['dif_p3_ciot_out'] = 0
df['dif_p3_bpc_out'] = 0
```

```python
for i in range (len(df)-1):
  df['dif_p2_ciot_in'].iloc[i] = df['p2_cont_iot_in '].iloc[i+1 ] -df['p2_cont_iot_in '].iloc[i]
  df['dif_p2_bpc_in'].iloc[i] = df['p2_bpc_in'].iloc[i+1 ] -df['p2_bpc_in'].iloc[i]
  df['dif_p2_ciot_out'] =       df['p2_cont_iot_out'].iloc[i+1 ] -df['p2_cont_iot_out'].iloc[i]
  
```

```python
for j in range (len(df)-1):
  df['dif_p3_ciot_in'].iloc[j] = df['p3_cont_iot_in '].iloc[j+1 ] -df['p3_cont_iot_in '].iloc[j]
  df['dif_p3_bpc_in'].iloc[j] = df['p3_bpc_in'].iloc[j+1 ] -df['p3_bpc_in'].iloc[j]
  df['dif_p3_ciot_out'].iloc[j] =       df['p3_cont_iot_out'].iloc[j+1 ] -df['p3_cont_iot_out'].iloc[j]
```

```python
for w in range (len(df)-1):
  df['dif_p2_bpc_out'].iloc[w] = df['p2_bpc_out'].iloc[w+1 ] -df['p2_bpc_out'].iloc[w]
  df['dif_p3_bpc_out'].iloc[w] = df['p3_bpc_out'].iloc[w+1 ] -df['p3_bpc_out'].iloc[w]
```
